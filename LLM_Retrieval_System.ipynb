```python
# ======================
# 1. Install Dependencies
# ======================
!pip install -q langchain langchain-community langchain-anthropic faiss-cpu pypdf pandas pytesseract pdf2image pillow matplotlib tqdm sentence-transformers tabulate

# Install OCR dependencies
!apt-get update -qq && apt-get install -qq tesseract-ocr poppler-utils

# ======================
# 2. Mount Google Drive
# ======================
from google.colab import drive
drive.mount('/content/drive')

# ======================
# 3. File System Navigator Class
# ======================
import time
import os
import json
import pickle
import pandas as pd
import numpy as np
from PIL import Image
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
from datetime import datetime
import pytesseract
from pdf2image import convert_from_path
from tabulate import tabulate

# Import LangChain components
from langchain.document_loaders import (
    PyPDFLoader,
    TextLoader,
    CSVLoader,
    UnstructuredMarkdownLoader,
    UnstructuredImageLoader,
    JSONLoader
)
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain_anthropic import ChatAnthropic
from langchain.schema import Document
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.retrievers.multi_query import MultiQueryRetriever
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

class FileSystemNavigator:
    def __init__(self,
                 root_dir,
                 index_dir=None,
                 anthropic_api_key=None,
                 chunk_size=3000,
                 chunk_overlap=500,
                 embedding_model="sentence-transformers/all-MiniLM-L6-v2",
                 llm_model="claude-3-sonnet-20240229",
                 max_file_size_mb=20,
                 verbose=True):
        """
        Initialize the FileSystemNavigator.

        Args:
            root_dir: Root directory to crawl
            index_dir: Directory to save/load index files
            anthropic_api_key: API key for Claude
            chunk_size: Size of text chunks for embedding
            chunk_overlap: Overlap between chunks
            embedding_model: Model to use for embeddings
            llm_model: Claude model to use
            max_file_size_mb: Maximum file size to process
            verbose: Whether to print detailed logs
        """
        self.root_dir = root_dir
        self.index_dir = index_dir or os.path.join(root_dir, "_index")
        self.anthropic_api_key = anthropic_api_key
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.embedding_model_name = embedding_model
        self.llm_model = llm_model
        self.max_file_size_mb = max_file_size_mb
        self.verbose = verbose

        # Create index directory if it doesn't exist
        os.makedirs(self.index_dir, exist_ok=True)

        # Initialize text splitter
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=self.chunk_size,
            chunk_overlap=self.chunk_overlap,
            length_function=len
        )

        # Initialize embeddings
        self.embeddings = HuggingFaceEmbeddings(
            model_name=self.embedding_model_name,
            model_kwargs={"device": "cuda" if os.environ.get("CUDA_VISIBLE_DEVICES") else "cpu"}
        )

        # Initialize LLM
        if self.anthropic_api_key:
            os.environ["ANTHROPIC_API_KEY"] = self.anthropic_api_key
            self.llm = ChatAnthropic(
                model=self.llm_model,
                temperature=0.1,
                max_tokens=1500
            )

        # File type handlers
        self.file_handlers = {
            ".pdf": self._process_pdf,
            ".txt": self._process_text,
            ".csv": self._process_csv,
            ".json": self._process_json,
            ".md": self._process_markdown,
            ".jpg": self._process_image,
            ".jpeg": self._process_image,
            ".png": self._process_image
        }

        # Stats
        self.stats = {
            "total_files": 0,
            "processed_files": 0,
            "skipped_files": 0,
            "error_files": 0,
            "total_chunks": 0,
            "file_types": {}
        }

        self.log("FileSystemNavigator initialized")

    def log(self, message):
        """Print log message if verbose is enabled"""
        if self.verbose:
            print(f"[{datetime.now().strftime('%H:%M:%S')}] {message}")

    def _process_pdf(self, file_path):
        """Process PDF file"""
        try:
            loader = PyPDFLoader(file_path)
            docs = loader.load()

            # If OCR is needed (no text extracted), use OCR
            if not docs or all(not doc.page_content.strip() for doc in docs):
                self.log(f"No text found in PDF, using OCR: {file_path}")
                return self._process_pdf_with_ocr(file_path)

            return docs
        except Exception as e:
            self.log(f"Error processing PDF {file_path}: {str(e)}")
            return []

    def _process_pdf_with_ocr(self, file_path):
        """Process PDF file using OCR"""
        try:
            # Convert PDF to images
            images = convert_from_path(file_path)
            docs = []

            for i, image in enumerate(images):
                # Extract text using OCR
                text = pytesseract.image_to_string(image)
                if text.strip():
                    docs.append(Document(
                        page_content=text,
                        metadata={"source": file_path, "page": i+1, "ocr": True}
                    ))

            return docs
        except Exception as e:
            self.log(f"Error processing PDF with OCR {file_path}: {str(e)}")
            return []

    def _process_text(self, file_path):
        """Process text file"""
        try:
            loader = TextLoader(file_path)
            return loader.load()
        except Exception as e:
            self.log(f"Error processing text file {file_path}: {str(e)}")
            return []

    def _process_csv(self, file_path):
        """Process CSV file"""
        try:
            # For large CSVs, read in chunks
            docs = []
            df = pd.read_csv(file_path)

            # Add metadata about columns
            columns_info = f"CSV Columns: {', '.join(df.columns.tolist())}"

            # Create a document with column information
            docs.append(Document(
                page_content=columns_info,
                metadata={"source": file_path, "content_type": "csv_schema"}
            ))

            # Add summary statistics for numeric columns
            numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
            if numeric_cols:
                stats_df = df[numeric_cols].describe().transpose()
                stats_text = f"CSV Statistics:\n{tabulate(stats_df, headers='keys', tablefmt='pipe')}"
                docs.append(Document(
                    page_content=stats_text,
                    metadata={"source": file_path, "content_type": "csv_stats"}
                ))

            # Convert to string in chunks to avoid memory issues
            max_rows = 1000
            for i in range(0, len(df), max_rows):
                chunk = df.iloc[i:i+max_rows]
                chunk_text = chunk.to_string(index=False)
                docs.append(Document(
                    page_content=chunk_text,
                    metadata={
                        "source": file_path,
                        "content_type": "csv_data",
                        "row_range": f"{i}-{min(i+max_rows, len(df))}"
                    }
                ))

            return docs
        except Exception as e:
            self.log(f"Error processing CSV file {file_path}: {str(e)}")
            return []

    def _process_json(self, file_path):
        """Process JSON file"""
        try:
            with open(file_path, 'r') as f:
                data = json.load(f)

            # Convert JSON to string representation
            json_str = json.dumps(data, indent=2)

            # Create document
            doc = Document(
                page_content=json_str,
                metadata={"source": file_path, "content_type": "json"}
            )

            return [doc]
        except Exception as e:
            self.log(f"Error processing JSON file {file_path}: {str(e)}")
            return []

    def _process_markdown(self, file_path):
        """Process Markdown file"""
        try:
            loader = UnstructuredMarkdownLoader(file_path)
            return loader.load()
        except Exception as e:
            self.log(f"Error processing Markdown file {file_path}: {str(e)}")
            return []

    def _process_image(self, file_path):
        """Process image file with OCR"""
        try:
            # Extract text using OCR
            image = Image.open(file_path)
            text = pytesseract.image_to_string(image)

            if text.strip():
                return [Document(
                    page_content=text,
                    metadata={"source": file_path, "content_type": "image_ocr"}
                )]
            else:
                return []
        except Exception as e:
            self.log(f"Error processing image file {file_path}: {str(e)}")
            return []

    def crawl_and_index(self, force_reindex=False):
        """
        Crawl the file system and create/update the index

        Args:
            force_reindex: If True, reindex all files even if they're already indexed
        """
        index_path = os.path.join(self.index_dir, "faiss_index")
        metadata_path = os.path.join(self.index_dir, "metadata.pkl")

        # Check if index already exists
        if os.path.exists(index_path) and os.path.exists(metadata_path) and not force_reindex:
            self.log(f"Loading existing index from {index_path}")
            self.vectorstore = FAISS.load_local(index_path, self.embeddings, allow_dangerous_deserialization=True)


            with open(metadata_path, 'rb') as f:
                self.metadata = pickle.load(f)

            self.log(f"Loaded index with {len(self.vectorstore.index_to_docstore_id)} chunks")
            return

        # Start fresh indexing
        self.log(f"Starting to crawl and index {self.root_dir}")
        start_time = time.time()

        all_docs = []
        processed_files = set()

        # Walk through the directory
                # --- Limit the number of files processed for testing ---
        # Change this to the number of files you want to process for testing
        N = 35
        # Collect all file paths in the directory and subdirectories
        all_files = []
        for root, _, files in os.walk(self.root_dir):
            for file in files:
                file_path = os.path.join(root, file)
                all_files.append(file_path)

        # Only keep the first N files
        all_files = all_files[:N]

        # Now process only these N files
        for file_path in tqdm(all_files, desc="Processing files"):
            # Skip index directory
            if self.index_dir in file_path:
                continue

            # Get file extension
            _, ext = os.path.splitext(file_path.lower())

            # Update stats
            self.stats["total_files"] += 1
            self.stats["file_types"][ext] = self.stats["file_types"].get(ext, 0) + 1

            # Skip unsupported file types
            if ext not in self.file_handlers:
                self.stats["skipped_files"] += 1
                continue

            # Skip files that are too large
            file_size_mb = os.path.getsize(file_path) / (1024 * 1024)
            if file_size_mb > self.max_file_size_mb:
                self.log(f"Skipping large file: {file_path} ({file_size_mb:.2f} MB)")
                self.stats["skipped_files"] += 1
                continue

            # Process the file
            try:
                self.log(f"Processing {file_path}")
                docs = self.file_handlers[ext](file_path)

                if docs:
                    # Split into chunks
                    chunks = self.text_splitter.split_documents(docs)
                    all_docs.extend(chunks)
                    processed_files.add(file_path)

                    self.stats["processed_files"] += 1
                    self.stats["total_chunks"] += len(chunks)

                    self.log(f"✅ Added {len(chunks)} chunks from {file_path}")
                else:
                    self.stats["skipped_files"] += 1
                    self.log(f"⚠️ No content extracted from {file_path}")
            except Exception as e:
                self.stats["error_files"] += 1
                self.log(f"❌ Error processing {file_path}: {str(e)}")

        # Create vector store
        if all_docs:
            self.log(f"Creating vector store with {len(all_docs)} chunks")

            # Process in batches to avoid memory issues
            batch_size = 1000
            vectorstore = None

            for i in tqdm(range(0, len(all_docs), batch_size), desc="Creating embeddings"):
                batch = all_docs[i:i+batch_size]
                batch_store = FAISS.from_documents(batch, self.embeddings)

                if vectorstore is None:
                    vectorstore = batch_store
                else:
                    vectorstore.merge_from(batch_store)

            self.vectorstore = vectorstore

            # Save index
            self.vectorstore.save_local(index_path)

            # Save metadata
            self.metadata = {
                "indexed_at": datetime.now().isoformat(),
                "root_dir": self.root_dir,
                "total_chunks": len(all_docs),
                "processed_files": list(processed_files),
                "stats": self.stats
            }

            with open(metadata_path, 'wb') as f:
                pickle.dump(self.metadata, f)

            self.log(f"✅ Indexing completed in {time.time() - start_time:.2f} seconds")
            self.log(f"✅ Processed {self.stats['processed_files']} files with {self.stats['total_chunks']} chunks")
            self.log(f"✅ Index saved to {index_path}")
        else:
            self.log("❌ No documents were processed. Check file types and permissions.")


    def query(self, question, k=5, use_multi_query=True):
        """
        Query the index with a natural language question

        Args:
            question: The question to ask
            k: Number of chunks to retrieve
            use_multi_query: Whether to use multi-query retrieval

        Returns:
            The answer from the LLM
        """
        if not hasattr(self, 'vectorstore'):
            self.log("No index found. Please run crawl_and_index() first.")
            return "Error: No index found. Please index your files first."

        self.log(f"Querying: {question}")

        # Get retriever
        if use_multi_query:
            # Use multi-query retrieval for better results
            retriever = MultiQueryRetriever.from_llm(
                retriever=self.vectorstore.as_retriever(search_kwargs={"k": k}),
                llm=self.llm
            )
        else:
            # Use standard retrieval
            retriever = self.vectorstore.as_retriever(search_kwargs={"k": k})

        # Retrieve relevant documents
        docs = retriever.get_relevant_documents(question)

        if not docs:
            return "I couldn't find any relevant information in the indexed files."

        # Format documents for context
        context = "\n\n".join([
            f"Document {i+1} (Source: {doc.metadata.get('source', 'Unknown')})\n{doc.page_content}"
            for i, doc in enumerate(docs)
        ])

        # Create prompt template
        prompt = ChatPromptTemplate.from_template("""
        You are a helpful assistant that answers questions based on the provided context from files.

        Context:
        {context}

        Question: {question}

        Answer the question based only on the provided context. If the context doesn't contain the information needed to answer the question, say "I don't have enough information to answer this question."

        Include relevant file sources in your answer. Be concise but thorough.
        """)

        # Create chain
        chain = (
            {"context": lambda x: context, "question": lambda x: question}
            | prompt
            | self.llm
            | StrOutputParser()
        )

        # Execute chain
        answer = chain.invoke({})

        return answer

    def display_stats(self):
        """Display statistics about the indexed files"""
        if not hasattr(self, 'metadata'):
            self.log("No index statistics available. Please run crawl_and_index() first.")
            return

        stats = self.metadata["stats"]

        print("\n" + "="*50)
        print("📊 File System Navigator Statistics")
        print("="*50)
        print(f"📁 Total files found: {stats['total_files']}")
        print(f"✅ Files processed: {stats['processed_files']}")
        print(f"⚠️ Files skipped: {stats['skipped_files']}")
        print(f"❌ Files with errors: {stats['error_files']}")
        print(f"📄 Total chunks indexed: {stats['total_chunks']}")
        print("\n📋 File types:")

        # Sort file types by count
        sorted_types = sorted(stats["file_types"].items(), key=lambda x: x[1], reverse=True)
        for ext, count in sorted_types:
            print(f"   {ext}: {count}")

        print("="*50)

        # Plot file type distribution
        plt.figure(figsize=(10, 6))
        extensions = [ext for ext, _ in sorted_types if ext in self.file_handlers]
        counts = [count for ext, count in sorted_types if ext in self.file_handlers]

        if extensions and counts:
            plt.bar(extensions, counts)
            plt.title('Distribution of Indexed File Types')
            plt.xlabel('File Extension')
            plt.ylabel('Count')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

# ======================
# 4. Main Application
# ======================

def main():
    # Set your API key
    ANTHROPIC_API_KEY = "fake_key_here"

    # Get root directory from user
    print("Enter the path to the directory you want to index (or press Enter to use the default):")
    user_dir = input().strip()

    # Default to a folder in Google Drive if no input
    ROOT_DIR = user_dir if user_dir else "/content/drive/MyDrive/file_system_navigator_data"

    # Create directory if it doesn't exist
    os.makedirs(ROOT_DIR, exist_ok=True)

    # Initialize navigator
    navigator = FileSystemNavigator(
        root_dir=ROOT_DIR,
        anthropic_api_key=ANTHROPIC_API_KEY,
        verbose=True
    )

    # Ask if user wants to reindex
    print("\nDo you want to reindex all files? (y/n)")
    reindex = input().strip().lower() == 'y'

    # Crawl and index
    navigator.crawl_and_index(force_reindex=reindex)

    # Display stats
    navigator.display_stats()

    # Interactive query loop
    print("\n" + "="*50)
    print("🔍 File System Navigator Query Interface")
    print("="*50)
    print("Type your questions below. Type 'exit' to quit.")
    print("="*50)

    while True:
        print("\nQuestion: ", end="")
        question = input().strip()

        if question.lower() in ['exit', 'quit']:
            break

        if not question:
            continue

        start_time = time.time()
        answer = navigator.query(question)
        query_time = time.time() - start_time

        print("\n" + "="*80)
        print(f"Answer (retrieved in {query_time:.2f} seconds):")
        print("="*80)
        print(answer)
        print("="*80)

if __name__ == "__main__":
    main()
```
